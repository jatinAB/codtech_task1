```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset from seaborn's built-in datasets
df = sns.load_dataset('iris')
```

```python
# Display the first few rows of the dataset
print(df.head())

# Get basic information about the dataset
print(df.info())

# Get summary statistics of the dataset
print(df.describe())
```

```python
# Check for missing values in the dataset
print(df.isnull().sum())
```

```python
# Plot histograms for each feature
df.hist(figsize=(10, 8), bins=15)
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
# Pairplot to visualize relationships
sns.pairplot(df, hue="species")
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

```python
# Exclude the 'species' column as it's categorical
corr_matrix = df.drop(columns=['species']).corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of Feature Correlations")
plt.show()
```

```python
# Boxplot for each feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplots of Iris Features")
plt.show()
```
